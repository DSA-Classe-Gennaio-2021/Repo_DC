```python
import numpy as np
import pandas as pd
ds = pd.read_csv('weatherAUS.csv')
ds = ds.drop(['Evaporation', 'Sunshine','Cloud9am','Cloud3pm'], axis='columns')
```

```python
ds = ds.dropna()
ds.reset_index()
ds
```

```python
#scaled numbers
X_num = ds.drop(['Date','Location','WindGustDir','WindDir9am','WindDir3pm','RainToday','RainTomorrow'], axis = 'columns')
from sklearn import preprocessing 
scaled_num = preprocessing.MinMaxScaler().fit_transform(X_num)
cols = X_num.columns
scaled_num = pd.DataFrame(data= scaled_num , columns= cols)
```

```python
#scale categorical
X_cat = ds.drop(['MinTemp','MaxTemp','Rainfall','WindGustSpeed','WindSpeed9am','WindSpeed3pm','Humidity9am','Humidity3pm','Pressure9am','Pressure3pm','Temp9am','Temp3pm'], axis = 'columns')
X_cat
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X_cat.Location = le.fit_transform(X_cat.Location)
X_cat.WindGustDir = le.fit_transform(X_cat.WindGustDir)
X_cat.WindDir9am = le.fit_transform(X_cat.WindDir9am)
X_cat.WindDir3pm = le.fit_transform(X_cat.WindDir3pm)
X_cat.RainToday = le.fit_transform(X_cat.RainToday)
X_cat.RainTomorrow = le.fit_transform(X_cat.RainTomorrow)
X_cat
```

```python
X_compl = pd.concat([X_cat,X_num], axis= 1)
X_compl = X_compl.set_index('Date')
X_compl
```

```python
x_train = X_compl.drop(['RainTomorrow'], axis='columns').head(90340)
y_train = X_compl.RainTomorrow.head(90340)
x_test = X_compl.drop(['RainTomorrow'], axis='columns')[90340:]
y_test = X_compl.RainTomorrow[90340:]
```

```python
from sklearn import tree
tree = tree.DecisionTreeClassifier()
tree.fit(x_train,y_train)
```

```python
tree.score(x_train,y_train)
```

```python
tree.score(x_test, y_test)
```

```python
from sklearn.linear_model import LogisticRegression
logr = LogisticRegression(solver='lbfgs',max_iter=1000).fit(x_train,y_train)
logr
```

```python
logr.score(x_train,y_train)
```

```python
logr.score(x_test,y_test)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_compl.drop(['RainTomorrow'], axis='columns'), X_compl.RainTomorrow, test_size=0.25)
```

```python
lr = LogisticRegression(solver='lbfgs',max_iter=1000)
lr.fit(X_train,Y_train)
lr.score(X_test,Y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=40)
rf.fit(X_train,Y_train)
rf.score(X_test,Y_test)
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=10)
kf
```

```python
def get_score(model,X_train,X_test,Y_train,Y_test):
    model.fit(X_train,Y_train)
    return model.score(X_test,Y_test)
```

```python
print ("lr score: ",get_score(LogisticRegression(solver='lbfgs',max_iter=1000),X_train,X_test,Y_train,Y_test))
print ("rf score: ",get_score(RandomForestClassifier(n_estimators=40),X_train,X_test,Y_train,Y_test))
print ("tree: ",get_score(tree,X_train,X_test,Y_train,Y_test))
```

```python
from sklearn.model_selection import StratifiedKFold
kf = StratifiedKFold(n_splits=10)
```
