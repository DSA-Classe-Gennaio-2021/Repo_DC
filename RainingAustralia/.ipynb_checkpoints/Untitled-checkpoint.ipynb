```python
#scaled numbers
X_num = dff.drop(['Date','Location','WindGustDir','WindDir9am','WindDir3pm','RainToday','RainTomorrow'], axis = 'columns')
from sklearn import preprocessing 
scaled_num = preprocessing.MinMaxScaler().fit_transform(X_num)
cols = X_num.columns.values
x_num = pd.DataFrame(data = scaled_num, columns=cols)
x_num
```

```python
x_str = dff.drop(['Date','MinTemp','MaxTemp','Rainfall','WindGustSpeed','WindSpeed9am','WindSpeed3pm','Humidity9am','Humidity3pm','Pressure9am','Pressure3pm','Temp9am','Temp3pm'], axis = 'columns')
x_str
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x_str.Location = le.fit_transform(x_str.Location)
x_str.WindGustDir = le.fit_transform(x_str.WindGustDir)
x_str.WindDir9am = le.fit_transform(x_str.WindDir9am)
x_str.WindDir3pm = le.fit_transform(x_str.WindDir3pm)
x_str.RainToday = le.fit_transform(x_str.RainToday)
x_str
```
