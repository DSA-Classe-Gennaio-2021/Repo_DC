```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('train.csv')
df.head()
X_train= df
```

```python
4250*80/100 #meglio usare
```

```python
train_num = X_train.drop(['state','area_code','international_plan','voice_mail_plan','churn'],axis='columns')
```

```python
from sklearn import preprocessing
```

scaling numbers

```python
scaled_num = preprocessing.MinMaxScaler().fit_transform(train_num)
cols = train_num.columns.values
scaled_num = pd.DataFrame(data = scaled_num, columns = cols)
scaled_num
```

encoding categoricals

```python
train_str = X_train.drop(['account_length','number_vmail_messages','total_day_minutes','total_day_calls','total_day_charge','total_eve_minutes','total_eve_calls','total_eve_charge','total_night_minutes','total_night_calls','total_night_charge','total_intl_minutes','total_intl_calls','total_intl_charge','number_customer_service_calls'], axis = 'columns')
train_str
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
str_ft=train_str
str_ft.state = le.fit_transform(str_ft.state)
str_ft.area_code = le.fit_transform(str_ft.area_code)
str_ft.international_plan = le.fit_transform(str_ft.international_plan)
str_ft.voice_mail_plan = le.fit_transform(str_ft.voice_mail_plan)
str_ft.churn = le.fit_transform(str_ft.churn)
str_ft
```

```python
final_X_train = pd.concat([scaled_num,str_ft], axis=1)
final_X_train
```

```python
training = final_X_train.head(3400)
```

```python
X = final_X_train[3400:].drop(['churn'], axis='columns')
Y = final_X_train[3400:].churn
```

```python
final_X_train.describe()
```

```python
x = training.drop(['churn'], axis='columns')
```

```python
y = training.churn
```

```python
from sklearn import tree
tree = tree.DecisionTreeClassifier()
```

```python
tree.fit(x,y)
```

```python
tree.score(x,y)
```

```python
tree.score(X,Y)
```

```python
from sklearn import svm
```

```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(svm.SVC(gamma='auto'), {
    'C': [1,10,20],
    'kernel': ['rbf','linear']
}, cv=10, return_train_score=False)
clf.fit(x, y)
clf.cv_results_
```

```python
dd = pd.DataFrame(clf.cv_results_)
dd
```

```python
from sklearn.model_selection import RandomizedSearchCV
rs = RandomizedSearchCV(svm.SVC(gamma='auto'), {
        'C': [1,10,20],
        'kernel': ['rbf','linear']  
    }, 
    cv=5, 
    return_train_score=False, 
    n_iter=2
)
rs.fit(x,y)
pd.DataFrame(rs.cv_results_)[['param_C','param_kernel','mean_test_score']]
```
